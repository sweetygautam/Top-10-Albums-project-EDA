```python
fig, axes=plt.subplots(2,2,figsize=(15,10))
# histogram and densit of funcation, set title
sns.distplot(albums.Year,ax=axes[0,0]).set_title("Year distribution")

# set number of bins and color, set title
sns.distplot(albums.Total_sale, bins=50,color='purple',ax=axes[0,1]).set_title('Total_sale distribution')
# only histogram without density funcation,set title
sns.distplot(albums.Ranking, kde=False, color='blue',ax=axes[1,0]).set_title('Ranking distribution')
# only histogram without density funcation,set title
sns.distplot(albums['Tracks'], hist=False, color='red',ax=axes[1,1]).set_title('Traks distribution')
```
